####4) What was the temperature in Central Park at noon on Christmas Day, 1980? 

#####Tip: Explore the Forecast.io API documentation.
#####Tip: Looking for UNIX time? You don't have to get it via python, you can Google around for an online tool.


```python
# "https://api.forecast.io/forecast/f240f160fd8831b7a3534c50a2a0b61e/340.7833, 73.9667,1980-12-25T12:00:00-0400"
# 40.7833° N, 73.9667° W
# 1980-12-25T12:00:00-0400
#UNIX time:346593600
```

```python
import urllib2
response_C = urllib2.urlopen("https://api.forecast.io/forecast/f240f160fd8831b7a3534c50a2a0b61e/40.7833,73.9667,346593600")
#response_C
contents_C = response_C.read()
import json
data_C = json.loads(contents_C)
data_C
```

```python
data_C['currently']['temperature']
```
